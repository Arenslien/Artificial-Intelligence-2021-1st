```python
# pandas 활용 O H E 2

import pandas as pd
df = pd.DataFrame({
    'A': ['a', 'b', 'a'],
    'B': ['b', 'a', 'c']
})
one_hot = pd.get_dummies(df['B'])
df = df.drop('B', axis = 1)

df = df.join(one_hot)

df
```

```python
# One-encoding with Scikit-learn

import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

x1 = pd.DataFrame({'country':['a', 'b', 'z', 'e', 'd']})

# DataFrame 전체를 라벨인코딩(숫자로 변환) 한 후, one-hot-encoding 해야함

le = LabelEncoder()
x = x1.apply(le.fit_transform)
print(x)
print(type(x))
print()

# One hot encoding
encoder = OneHotEncoder()
x = encoder.fit_transform(x)
print(x)
print(type(x))
print()

x = x.toarray()
print(x)
print(type(x))
```

```python
# One-encoding with Scikit-learn

import pandas as pd
from sklearn.preprocessing import OneHotEncoder

x1 = pd.DataFrame({'country':['a', 'b', 'z', 'e', 'd']})



# One hot encoding
encoder = OneHotEncoder(categories = 'auto')
x = encoder.fit_transform(x1)
print(x)
print(type(x))
print()

x = x.toarray()
print(x)
print(type(x))
```

```python
# One-encoding with Scikit-learn

import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

x1 = pd.DataFrame({
    'A':['a', 'b', 'z', 'e', 'd'],
    'B':['x', 'y', 'x', 'y', 'x']
})

# LabelEncoding
le = LabelEncoder()
x = x1.apply(le.fit_transform)
print(x)
print(type(x), "\n")

# One Hot Encoding
encoder = OneHotEncoder(categories = "auto")
x2 = encoder.fit_transform(x)

print(x2)
print(type(x2), "\n")

x2 = x2.toarray()
print(x2)
print(type(x2), "\n")
```

```python
# One-encoding with Scikit-learn

import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

x1 = pd.DataFrame({
    'A': ['a', 'b', 'c', 'e', 'd'],
    'B': [3, 4, 5, 6, 7]
})

ct = ColumnTransformer([('one_hot_encoder', OneHotEncoder(), [0])], remainder = 'passthrough')



# LabelEncoding
le = LabelEncoder()
x1.A = le.fit_transform(x1.A)
print(x1)
print(type(x1), "\n")

# One Hot Encoding
x2 = ct.fit_transform(x1)

print(x2)
print(type(x2), "\n")
```
