```python
# Artificial Intelligence Quiz about week 4
# 60191686 정성훈

# URL 라이브러리
import urllib.request
import urllib.parse

url = "https://search.naver.com/search.naver"
values = {
    "query":"anno+domini",
    "sm":"tab_hty.top",
    "where":"nexearch",
    "oquery":"%EC%95%84%EB%85%B8%EB%8F%84%EB%AF%B8%EB%8B%88",
    "tqi":"hbuhsdp0JXVssavIYR4ssssssPV-044913",
}

params = urllib.parse.urlencode(values)
url = url + "?" + params

print("url = ", url, "\n")

# 다운로드
data = urllib.request.urlopen(url).read()
text = data.decode("utf-8")
print("text:", text)
```
