```python
# Scaling

import pandas as pd
from sklearn.preprocessing import scale, minmax_scale


x = pd.DataFrame({'col':[-3, -1, 1, 3, 5, 7, 9]})


# 평균 0, 분산을 이용해 정규화
x["scale"] = scale(x.col.astype(float))

# 0-1 사이의 값으로 정규화
x["minmax_scale"] = minmax_scale(x.col.astype(float))

x
```

```python
import pandas as pd
from sklearn.preprocessing import scale, minmax_scale

x = pd.DataFrame({"col":[-3, -1, 1, 3, 5, 7, 9]})

x["scale"] = scale(x.col.astype(float))
x["minmax_scale"] = minmax_scale(x.col.astype(float))

x.describe()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

dfTest = pd.DataFrame({
    'A': [14, 90.20, 90.95, 96.27, 91.21],
    'B': [103.02, 107.2, 110.35, 114.23, 114.68],
    'C': ['big', 'small', 'big', 'small', 'small']
})

dfTest[['A', 'B']] = scaler.fit_transform(dfTest[['A', 'B']])
dfTest
```

```python
# Nominal Attributes

import pandas as pd
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

encoding_data = ["paris", "paris", "tokyo", "amsterdam"]

le.fit(encoding_data)
print(le.classes_)
print(type(le.classes_), "\n")

data = le.transform(encoding_data)
print(data)
print(type(data), "\n")

original = le.inverse_transform(data)
print(original)
print(type(original))
```

```python
# Transform 예제

from sklearn import preprocessing
le = preprocessing.LabelEncoder()

str = []
for i in range(ord('a'), ord('z')+1, 1):
    str.append(chr(i))

le.fit(str)
data = le.transform(['a', 'd', 'f', 'z'])
print(data)
```

```python
# DataFrame 변환

import pandas as pd
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
df = pd.DataFrame({
    'A': ['a', 'b', 'b', 'c', 'a'],
    'B': ['x', 'y', 'x', 'y', 'x'],
})

data = df.apply(le.fit_transform)
print(data)
print(type(data))
```
