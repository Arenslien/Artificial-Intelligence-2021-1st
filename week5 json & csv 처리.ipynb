```python
import urllib.request as req
import json

json_str = req.urlopen("https://api.github.com/repositories").read()
output = json.loads(json_str)

for item in output:
    print(item["name"])
    print(item["full_name"])
    print(item["owner"]["login"])
    print()
```

```python
# JSON 형식으로 출력
import json
price = {
    "data": "2021-04-03",
    "price": {
        "apple": 80,
        "orange": 20,
        "banana": 40
    }
}

# Json 형식의 문자열로 변환, 즉, json.loads의 역 개념이다.
json_str = json.dumps(price)
print(json_str)

# 다시 파이썬 객체로 변환
data =json.loads(json_str)
print(data["data"])
print(data["price"]["apple"])
```
